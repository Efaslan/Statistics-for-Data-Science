# **Normal dağılıma sahipler mi?**

```python
from scipy.stats import anderson

#Sağlıklı insanlar için
results = anderson(np.array(le.fit_transform(dfs_sample.AlcoholDrinkers)))

print('Anderson-Darling Test Statistic:', results.statistic)
print('Kritik Değerler:', results.critical_values)
print('Signifikans Seviyeleri:', results.significance_level)

for i in range(len(results.critical_values)):
    sl, cv = results.significance_level[i], results.critical_values[i]
    if results.statistic < cv:
        print(f"{sl}% seviyesinde, veriler bu dağılıma uyuyor.")
    else:
        print(f"{sl}% seviyesinde, veriler bu dağılıma uymuyor.")

#Kalp Krizi geçirmişler için
from scipy.stats import anderson

resulth = anderson(np.array(le.fit_transform(dfh_sample.AlcoholDrinkers)))

print('\nAnderson-Darling Test Statistic:', resulth.statistic)
print('Kritik Değerler:', resulth.critical_values)
print('Signifikans Seviyeleri:', resulth.significance_level)

for i in range(len(resulth.critical_values)):
    sl, cv = resulth.significance_level[i], resulth.critical_values[i]
    if resulth.statistic < cv:
        print(f"{sl}% seviyesinde, veriler bu dağılıma uyuyor.")
    else:
        print(f"{sl}% seviyesinde, veriler bu dağılıma uymuyor.")
```

# **Oran Testi**

 *1. Durum Hasta olan gruptaki alkol kullanan oranı sağlıklı gruptakine eşittir.*

```python
from statsmodels.stats.proportion import proportions_ztest

count = np.array([dfh_sample.AlcoholDrinkers.value_counts()['Yes'], dfs_sample.AlcoholDrinkers.value_counts()['Yes']])
nobs = np.array([len(dfs_sample), len(dfh_sample)])

stat, p_value = proportions_ztest(count, nobs, alternative='two-sided')

print("Test İstatistiği:", stat)
print("p-değeri:", p_value)

if (p_value < alpha):
    print("Hipotezimiz yanlıştır.")
else:
    print("Hipotezimiz doğru.")
```


*2. Durum Hasta olan gruptaki alkol kullanan oranı sağlıklı gruptakinden büyüktür*




```python
stat, p_value = proportions_ztest(count, nobs, alternative='larger')

print("Test İstatistiği:", stat)
print("p-değeri:", p_value)

if (p_value < alpha):
    print("Hipotezimiz yanlıştır.")
else:
    print("Hipotezimiz doğru.")
```
